<h4 align='justify'>We will combine two dataset on the basis of "Country Code" column.</h4>

```python
df_final = pd.merge(df, df_country, on='Country Code', how='left')
df_final
```

```python
# To check data types
df_final.dtypes
```

```python
df_final.Country.value_counts()
```

```python
country_names = df_final.Country.value_counts().index
```

```python
country_val = df_final.Country.value_counts().values
```

```python
# pie chart
plt.pie(country_val[:3], labels = country_names[:3], autopct = '%1.1f%%')
```

<h4>Observation:India coutry has more Zomato transition as compair to United Kingdom and United states</h4>

```python
df_final.columns
```

```python
ratings = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
ratings
```

Observationas:
1. When rating is between 4.5-4.9 ----> Excellent
2. When Rating is between 4.0-4.4 ----. Very Good
3. When Rating is between 3.5-3.9 ----> Good
4. When Rating is between 2.5-3.4 ----> Average
5. When Rating is between 1.8-2.4 ----> Poor
6. When rating is 0.0 ----> Not Rated

```python
plt.figure(figsize=(12,6))
sns.barplot(x = "Aggregate rating", y ='Rating count', data = ratings)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y ='Rating count', hue = 'Rating color', data = ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

Observation:
1. Not rated count is very high. 
2. Maximum rating in between 2.5-3.5

```python
# count plot is basically used for oject/string/categorical varibales.
sns.countplot(x = 'Rating color', data = ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```
