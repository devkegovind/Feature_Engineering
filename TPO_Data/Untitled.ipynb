```python
y = [3.27,4.49, 3.69,3.49,3.82]
x = ['2017-18', '2018-19', '2019-20','2020-21','2021-22']
plt.figure(figsize=(9,7))
sns.set_context("talk")
sns.set_style("white")
ax = sns.barplot(x=x, y=y,  errwidth=0)
ax.bar_label(ax.containers[0])
plt.xlabel("Academic_Years", fontweight = 'bold')
plt.ylabel("Package(LPA)", fontweight = 'bold')
plt.title("Average Company Packages(LPA)", size = 20, fontweight='bold')
plt.tight_layout()
plt.savefig('avg1111.png', format="png", dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plot
import pandas as pd
# Make a data definition
_data = {'Higher_Study': [24, 7, 4, 20, 28],
        'Placed': [91, 36, 51, 107, 85]}
_df = pd.DataFrame(_data,columns=['Higher_Study', 'Placed'], index = ['2017-18', '2018-19', '2019-20', '2020-21', '2021-22'])
 
# Multiple horizontal bar chart
plt.figure(figsize=(7,5))
ax = _df.plot.bar( )
ax.set_facecolor('white')
plt.title("Higher_Study & Placed Students", size = 15, fontweight='bold')
plt.xlabel("Academic_Years", fontweight = 'bold', size=12)
plt.ylabel("No. of Students", fontweight = 'bold', size=12)
plt.xticks(size=12, rotation=0)
plt.yticks(size=12, rotation=0)
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('bar22.png', format="png", dpi=600)
plot.show()
```
