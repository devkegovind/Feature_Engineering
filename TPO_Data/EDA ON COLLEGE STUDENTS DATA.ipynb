```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df_tpo = pd.read_excel("tpo_data2.xlsx")
df_tpo
```

```python
df_tpo.shape
```

```python
df_tpo.drop('Program graduated from  ', axis =1, inplace = True)
```

```python
df_tpo
```

```python
df_tpo['Name'] = df_tpo['Name of student placed   / enrolling into higher education and contact details'].str.split(',').str[0]
```

```python
df_tpo['Contact'] = df_tpo['Name of student placed   / enrolling into higher education and contact details'].str.split(',').str[1]
```

```python
df_tpo['University'] = df_tpo['Name of the  employer with contact details / Name of institution joined'].str.split(',').str[0].str.split(';').str[0]
```

```python
df_tpo.rename(columns={"Pay package at appointment (In INR per annum) (applicable for students who got placement) / Name of program admitted to (applicable for students who progressed to higher education)" : "Education"}, inplace = True)
```

```python
df_tpo.columns
```

```python
df_tpo.drop([ 'Name of student placed   / enrolling into higher education and contact details',
       'Name of the  employer with contact details / Name of institution joined'], axis = 1, inplace = True)
```

```python
df_tpo
```

```python
df_tpo['University'] = df_tpo['University'].apply(lambda x : x.split(" ")[0])
```

```python
df_tpo.info()
```

```python
df_tpo['Education'].unique()
```

```python
df_tpo['University'] = df_tpo['University'].str.replace("Dr.", "Dr. BATU")
```

```python
df_tpo['Education'] = df_tpo['Education'].str.replace("M.Tech", "PG")
```

```python
char_to_remove = ['PGD','PGDCA','PG Diploma','PG-DAC']
col_to_clean = ['Education']
for item in char_to_remove:
    for col in col_to_clean:
        df_tpo[col] = df_tpo[col].str.replace(item, "PGD")
```

```python
df_tpo['Education'].unique()
```

```python
df_tpo.columns
```

```python
df_tpo.rename(columns={'Year ' : 'Year'}, inplace = True)
```

```python
df_tpo.columns
```

```python
df_tpo['Year'].value_counts().sort_values(ascending=False)
```

```python
df_tpo['Education'].value_counts().sort_values(ascending=False)
```

```python
ser = df_tpo.groupby(['Year'])['Education'].value_counts()
ser
```

```python
ser_df = pd.DataFrame(ser)
ser_df
```

```python
ser_df.rename(columns={"Education":"Count"}, inplace = True)
ser_df.to_excel("Summary_edu.xlsx")
```

```python
plt.title("Education_Wise Count of Higher_Study", size=20, fontweight='bold')
sns.set_theme(style="white")
sns.countplot(x=df_tpo['Education'])
plt.xlabel("Education_Types", fontweight = 'bold')
plt.ylabel("No. of Students", fontweight = 'bold')
plt.xticks(size=15, rotation=0)
plt.tight_layout()
#plt.savefig("Education_Wise Count of Higher_Study1.png", format = 'png', dpi = 600)
plt.show()
```

```python
df_tpo['University'].value_counts().sort_values(ascending=False)
```

```python
plt.title("Year-Wise Count of Higher_Study", size=20, fontweight='bold')
sns.set_theme(style="darkgrid")
sns.countplot(x=df_tpo['Year'])
plt.xticks(size=15, rotation=90)
plt.tight_layout()
plt.savefig("Year-Wise Count of Higher_Study.png")
plt.show()
```

```python
plt.figure(figsize=(35,25))
plt.title("University-Wise Count of Placed Students", size=20, fontweight='bold')
sns.set_theme(style="darkgrid")
sns.countplot(x='University', hue ='Year', data = df_tpo)
plt.xticks(size=15, rotation=90)
plt.tight_layout()
plt.savefig('University-Wise Count of Placed Students.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Higher Studies Count of Students Year Wise", size=15, fontweight='bold')
sns.set_theme(style="white")
ax = sns.countplot(x=df_tpo['Education'], hue =df_tpo['Year'])
plt.xticks(size=10, rotation=0)
plt.yticks(size=10, rotation=0)
plt.legend(loc='upper right')
plt.ylabel("No. of Students", fontweight = 'bold', size=12)
plt.xlabel("Eduction_Types", size=12, fontweight = 'bold')
plt.tight_layout()
plt.savefig('Education-Wise Count of Students2224.png', format = 'png', dpi = 600)
plt.show()
```
