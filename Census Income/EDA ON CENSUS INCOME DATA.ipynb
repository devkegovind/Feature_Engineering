```python
job_types = ['Private', 'Self-emp-not-inc', 'Local-gov', 'Self-emp-inc','State-gov']
sns.set_context("poster")
sns.set_style("darkgrid")
plt.figure(figsize=(40,30))
for i, job in enumerate(job_types):
    df3 = dje[dje['Job_Type'] == job].head()
    plt.subplot(4,2, i+1)
    sns.barplot(x = "HRA", y = "Job_Type", data = df3)
    plt.xlabel("Experience in Years")
    plt.ylabel("")
plt.tight_layout()
plt.subplots_adjust(hspace = .3)
plt.show()
```
